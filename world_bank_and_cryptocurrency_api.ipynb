```python
import numpy as np

def fibonacci_retracements(data, start_idx, end_idx):
    start_price = data['Close'][start_idx]
    end_price = data['Close'][end_idx]
    price_range = end_price - start_price

    levels = [0, 0.236, 0.382, 0.5, 0.618, 0.764]
    retracements = [start_price + price_range * level for level in levels]

    return pd.Series(retracements, index=[f"{level*100}%" for level in levels])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch Solana price data from Yahoo Finance
data_sol = yf.download("SOL-USD", start="2021-01-01", end="2023-04-30")

# Simple Moving Average (SMA)
def SMA(data, n):
    sma = data['Close'].rolling(window=n).mean()
    return sma

# Exponential Moving Average (EMA)
def EMA(data, n):
    ema = data['Close'].ewm(span=n, adjust=False).mean()
    return ema

# Momentum (MOM)
def MOM(data, n):
    mom = data['Close'].diff(n)
    return mom

# Rate of Change (ROC)
def ROC(data, n):
    m = data['Close'].diff(n - 1)
    n = data['Close'].shift(n - 1)
    roc = m / n
    return roc

# Average True Range (ATR)
def ATR(data, n):
    high = data['High'].values
    low = data['Low'].values
    close = data['Close'].values

    tr = np.zeros_like(close)
    tr[0] = high[0] - low[0]
    for i in range(1, len(close)):
        tr[i] = max(high[i] - low[i], abs(high[i] - close[i-1]), abs(low[i] - close[i-1]))

    atr = np.zeros_like(close)
    atr[0] = tr[0]
    for i in range(1, len(close)):
        atr[i] = (atr[i-1] * (n-1) + tr[i]) / n

    return atr

# Plot the indicators
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data_sol.index, data_sol['Close'], label='Price')
ax.plot(data_sol.index, SMA(data_sol, 20), label='SMA 20')
ax.plot(data_sol.index, EMA(data_sol, 20), label='EMA 20')
ax.plot(data_sol.index, MOM(data_sol, 14), label='Momentum 14')
ax.plot(data_sol.index, ROC(data_sol, 14), label='ROC 14')
ax.plot(data_sol.index, ATR(data_sol, 14), label='ATR 14')
ax.legend()
plt.show()
```

```python
#Plotting various technical indicators
#Moving Averages
def SMA(data,n):
  sma=data['Close'].rolling(window=n).mean()
  return sma

#Exponential Moving Average
def EMA(data,n) :
  ema = data['Close'].ewm(span=n,adjust=False).mean()
  return ema
#Momentum(MOM)
def MOM(data,n):
  mom = data['Close'].diff(n)
  return mom

#Rate of change (ROC)
def ROC(data,n):
  m = data['Close'].diff(n-1)
  n = data['Close'].shift(n-1)
  roc = m/n
  return roc

# Average True Range (ATR)
def ATR(data, n):
    high = data['High'].values
    low = data['Low'].values
    close = data['Close'].values

    trange = np.zeros_like(close)
    trange[0] = high[0] - low[0]
    for i in range(1, len(close)):
        trange[i] = max(high[i] - low[i], abs(high[i] - close[i-1]), abs(low[i] - close[i-1]))

    atr = np.zeros_like(close)
    atr[0] = trange[0]
    for i in range(1, len(close)):
        atr[i] = (atr[i-1] * (n-1) + trange[i]) / n #average of true ranges in the past plus the current one

    return atr

```
