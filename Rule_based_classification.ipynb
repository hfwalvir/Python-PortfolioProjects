#We divide customers into different segments A,B,C,D

```python
data.loc[:,'SEGMENT'] = pd.qcut(data['PRICE'],4,labels=['A','B','C','D'])
data.head(n=5)
```

SETTING THE CRITERIA

```python
agg_df.loc[(agg_df['AGE'] <=18),'AGE_CAT']= "0_18"
agg_df.loc[(agg_df['AGE']>18) & (agg_df['AGE']<24),'AGE_CAT']="19_23"
agg_df.loc[(agg_df['AGE']>=24)& (agg_df['AGE']<31),'AGE_CAT']= '24_30'
agg_df.loc[(agg_df['AGE']>=31)& (agg_df['AGE']<41),'AGE_CAT']= '31_40'
agg_df.loc[(agg_df['AGE']>=41)& (agg_df['AGE']<71),'AGE_CAT']= '41_70'
```

```python
agg_df.AGE_CAT.value_counts()
```

```python
agg_df["LEVEL_OF_CUSTOMER"] = agg_df['COUNTRY'].astype('str') + '_'+agg_df['SOURCE'].astype('str') + '_'+agg_df['SEX'].astype(str)+'_'+agg_df['AGE_CAT'].astype(str)
```

```python
del agg_df["COUNTRY"]
del agg_df["SOURCE"]
del agg_df["SEX"]
del agg_df["AGE"]
del agg_df["AGE_CAT"]
```
